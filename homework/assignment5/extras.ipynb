```python
def sort(vals):
    """Sorting Algorithm."""
    sorted_vals = []
    # max = vals[0]
    # vals = [6,17,18,18,100,4,3,0]
    # # for value, vector in zip(vals, vec):
    for index, value in enumerate(vals):
        if sorted_vals == [] or value < sorted_vals[-1]:
            sorted_vals.append(value)
        
        else:
            # sorted_vals_len = len(sorted_vals)
            not_appended = True
            for i in range(len(sorted_vals)):
                if value > sorted_vals[i] and not_appended:
                    sorted_vals = sorted_vals[0:i] + [value] + sorted_vals[i:len(sorted_vals)]
                    not_appended = False

    print(sorted_vals)
        
```
